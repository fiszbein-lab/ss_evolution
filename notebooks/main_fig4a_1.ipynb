```python
import csv
import glob
```

```python
hsap2mmus = dict()
```

```python
for fp in glob.iglob(f"data/homology/exons/chrom/*.csv"):
    with open(fp, 'r') as f:
        reader = csv.DictReader(f)
        
        for row in reader:
            hsap = row['anchor']
            mmus = row['mus_musculus']
            
            if mmus != "-":
               hsap2mmus[hsap] = mmus
```

```python
with (open(f"../data/homology/exons/exons.csv", 'r') as rf,
      open("../data/homology/hsap_mmus.csv", 'w') as wf):
    reader = csv.DictReader(rf)
    writer = csv.writer(wf)
    
    writer.writerow([
        "hsap_exon",
        "hsap_type",
        "mmus_exon",
        "mmus_type",
    ])
    
    for row in reader:
        hsap = row['hsap_exon']
        mmus = row['mus_musculus']
        
        hsap_type, hsap_exon = hsap.split("=")

        if hsap_exon in hsap2mmus:
            if mmus == "-":
                mmus_type, mmus_exon = "N", hsap2mmus[hsap_exon]
            else:
                if ";" in mmus:
                    mmus_type, mmus_exon = "M", hsap2mmus[hsap_exon]
                else:
                    mmus_type, mmus_exon = mmus.split("=")
            
            writer.writerow([
                hsap_exon,
                hsap_type,
                mmus_exon,
                mmus_type,
            ])
```
