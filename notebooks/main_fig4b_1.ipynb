```python
import csv
import glob
```

```python
hsap2mmus = dict()
```

```python
for fp in glob.iglob("data/homology/exons/chrom/*.csv"):
    with open(fp, 'r') as f:
        reader = csv.DictReader(f)
        
        for row in reader:
            hsap = row['anchor']
            mmus = row['mus_musculus']

            if mmus != "-":
                hsap2mmus[hsap] = mmus
```

```python
with (open("../data/homology/exons/exons.csv", 'r') as rf,
      open("../data/homology/exons/hsap_mmus.csv", 'w') as wf):
    reader = csv.DictReader(rf)
    writer = csv.writer(wf)
    
    writer.writerow([
        "hsap_exon",
        "hsap_exon_type",
        "mmus_exon",
        "mmus_exon_type",
    ])
    
    for row in reader:
        hsap = row['hsap_exon']
        mmus = row['mus_musculus']
        
        hsap_type, hsap_exon = hsap.split("=")

        if hsap_exon in hsap2mmus:
            if mmus == "-":
                # We have a mouse alignment for the human exon, but the
                # alginment didn't overlap an annotated mouse exon.
                mmus_type, mmus_exon = "NE", hsap2mmus[hsap_exon]
            else:
                # We have a mouse alignment for the human exon, and the
                # alginment did overlap an annotated mouse exon.
                mmus_type, mmus_exon = mmus.split("=")
            
            writer.writerow([
                hsap_exon,
                hsap_type,
                mmus_exon,
                mmus_type,
            ])
```
