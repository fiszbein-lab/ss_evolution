```python
import csv
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tree_no = {'0': 0, '1': 0, '2+': 0}
```

```python
with open("../data/homology/ss/out/tree_no_dist.txt", 'r') as f:
    for row in f:
        num = int(row.strip("\n"))

        if num < 2:
            tree_no[str(num)] += 1
        else:
            tree_no['2+'] += 1
```

```python
percentages = [tree_no[num] / sum(tree_no.values()) * 100 for num in ['0', '1', '2+']]

_ = plt.pie(percentages, labels=["0", "1", "2+"], autopct='%1.1f%%')
plt.savefig("figs/supp1a.pdf")
```

```python
tree_sizes = list()
```

```python
with open("../data/homology/ss/out/origin.csv", 'r') as f:
    reader = csv.reader(f)
    next(reader)

    for row in reader:
        *_, tree_size, _ = row

        tree_size = int(tree_size)
        tree_sizes.append(tree_size)
```

```python
sns.histplot(tree_sizes, bins=20, element='step')

_ = plt.xlim(0, 63)
_ = plt.ylim(0, 5000)

_ = plt.axvline(x=3, color='red')

plt.savefig("figs/supp1b.pdf")
```
